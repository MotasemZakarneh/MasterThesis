Use this, to read the data from the csv and plot it in python

```python
import csv as creader
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

method 1 using csv reader

```python
file = open("Data/Tester/BS_DOS.csv")
file_reader = creader.reader(file)
headers = next(file_reader)
# target data form is :
# [[x0,y0],[x1,y1]........[x9,y9]]

python_arr = []
i = 10
for row in file_reader:
    if row[0] == "" or row[0] == "--" or row[1] == "" or row[1] == "--":
        continue
    a = [row[0], row[1]]
    python_arr.append(a)

file.close()

plot_data = np.array(python_arr,dtype=float) #remember to cast for float
#print(plot_data.shape) (7128,2) [ many elements, each element is 2 parts ]
ax = plot_data[:,0]#get all inner arrays, and only choose the first element of them
by = plot_data[:,1]#get all inner arrays, and only choose the second element of them

ax = np.array(ax)
by = np.array(by)

plt.plot(ax,by)
plt.show()
```

Method 2 using Pandas

```python
df = pd.read_csv("Data/Tester/BS_DOS.csv")
#Get the Data that we need
pair = df.get(["Ax", "By"]) #DataFrame class

#split data
ax = pair.get("Ax") #Probably A Sesries?
#filter bad data
ax = ax.replace(["--",""], ["0","0"])
#cast to float
ax = ax.astype(float)

#split data
by = pair.get("By")
#filter bad data
by = by.replace(["--",""], ["0","0"])
#cast to float
by = by.astype(float)

#get np array
ax = np.array(ax)
by = np.array(by)

plt.plot(ax,by)
#plt.ylabel('some numbers')
plt.show()
```
